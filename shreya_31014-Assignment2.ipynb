<h3> Normal curves for class test result </h3>

```python
plt.hist(physics,bins=10)
plt.title('Histogram for physics score')
plt.xlabel('Physics marks')
plt.legend()
plt.show()
```

```python
plt.close()
```

```python
std=np.std(physics,ddof=1)
mean=np.mean(physics)
```

```python
import matplotlib.pyplot as plt

domain=np.linspace(np.min(physics),np.max(physics),1000)
plt.plot(domain,norm.pdf(domain,mean,std))
plt.title('Normal distribution curve')
plt.xlabel('Physics marks')
plt.ylabel('PDF')
plt.legend()
plt.show()
```

```python
plt.close()
```

```python
mean1=np.mean(chemistry)
std1=np.std(chemistry,ddof=1)
```

```python
import matplotlib.pyplot as plt

domain=np.linspace(np.min(chemistry),np.max(chemistry),1000)
plt.plot(domain,norm.pdf(domain,mean1,std1))
plt.title('Normal distribution curve')
plt.xlabel('Chemistry score')
plt.ylabel('PDF')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

domain=np.linspace(np.min(chemistry),np.max(chemistry),1000)
plt.plot(domain,norm.pdf(domain,mean1,std1))
plt.hist(chemistry,edgecolor='black',alpha=.5,density=True)
plt.title('Normal distribution curve')
plt.xlabel('Chemistry score')
plt.ylabel('PDF')
plt.legend()
plt.show()
```

<h3> To find skewness </h3>

```python
from scipy.stats import skew
```

```python
print('Skewness for physics marks :',skew(physics.dropna()))   
# marks in physics are slightly negatively skewed indicating curve is slightly elongated at left tail
```

```python
print('Skewness for chemistry marks :',skew(chemistry.dropna()))
# marks in chemistry are slightly positively skewed indicating curve is slightly elongated at right tail
```

<h3>To find kurtosis: </h3>

```python
from scipy.stats import kurtosis
```

```python
print('Kurtosis for physics marks :',kurtosis(physics.dropna()))
# since kurtosis is approximately 0, curve is mesokurtic
```

```python
print('Kurtosis for chemistry marks :',kurtosis(chemistry.dropna()))
# since kurtosis is approximately 0, curve is mesokurtic
```
