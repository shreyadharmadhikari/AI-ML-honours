```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn
from scipy.stats import pearsonr
```

```python
df=pd.read_csv(r"C:\Users\Suhas\Desktop\AI and ML\birthwt.csv")    #Read CSV file
```

```python
age=df['age']                #Store all the values of column age in birthwt.csv into variable age
bwt=df['bwt']                #Store all the values of column bwt in birthwt.csv into variable bwt
```

```python
corr, _ = pearsonr(age, bwt)         #Find pearson's correlation coefficient using pearsonr(x,y) function
print('Pearsons correlation coefficient : %.3f' % corr)
```

```python
plt.scatter(age,bwt)                 #Plot a scatter plot between mother's age in years and birthweight of baby in grams
plt.xlabel("mother's age in years")  #label for the X-axis 
plt.ylabel('birthweight in grams')   #label for the Y-axis
plt.title('plot between mothers age and birthweight' )  #Title of the graph
```

```python
from scipy import stats
```

```python
slope,intercept,r,p,std_err=stats.linregress(age,bwt)    #To find correlation coefficient between age and bwt
```

```python
print('Correlation coefficient:',r)                      #To print correlation coefficient
```

```python
#Thus pearson correlation coefficient is approximately same as the correlation coefficient calculated with method linregress()
```

```python
from scipy.stats import spearmanr
```

```python
#To find correlation coefficient using Numpy between age and bwt
corc=np.corrcoef(age,bwt)
print('Numpy Correlation coefficient:',corc)  #r is same as pearson's correlation coefficient
```

```python
#using pandas
age.corr(bwt)  #pearson's r
```

```python
#using pandas
bwt.corr(age)  #pearson's r
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
slope, intercept, r, p, stderr = stats.linregress(age, bwt)        
```

```python
line = f'Regression line: y={intercept:.2f}+{slope:.2f}x, r={r:.2f}'
line
```

```python
fig, ax = plt.subplots()                  #Plot of linear regression between age and bwt 
ax.plot(age, bwt, linewidth=0, marker='s', label='Data points')
ax.plot(age, intercept + slope * age, label=line)
ax.set_xlabel("mother's age in years")
ax.set_ylabel('birthweight in grams')
ax.legend(facecolor='white')
plt.show()
```
